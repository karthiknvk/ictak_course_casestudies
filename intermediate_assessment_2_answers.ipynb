```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
```

```python
X=pd.DataFrame(X,columns=X_columns)
X
```

<h4>SPLITTING DATA</h4>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42,stratify=y)
```

<h4>TRAINING MODEL (with fine tuning)</h4>

```python
from sklearn.metrics import f1_score,classification_report,confusion_matrix
```

#LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression

lr_model=LogisticRegression()
lr_model.fit(X_train,y_train)
lr_model_y_pred=lr_model.predict(X_test)
print("f1 score : ",f1_score(y_test,lr_model_y_pred))
print(classification_report(y_test,lr_model_y_pred))
print(confusion_matrix(y_test,lr_model_y_pred))
```

#KNN 

```python
from sklearn.neighbors import KNeighborsClassifier

f1_score_list=[]
for i in range(1,15):
  knn_model=KNeighborsClassifier(n_neighbors=i,metric='minkowski')
  knn_model.fit(X_train,y_train)
  knn_model_y_pred=knn_model.predict(X_test)
  f1_score_list.append(f1_score(y_test,knn_model_y_pred))
```

```python
plt.plot(np.arange(1,15),f1_score_list,'-o')
plt.show()
```

```python
knn_model=KNeighborsClassifier(n_neighbors=3,metric='minkowski')
knn_model.fit(X_train,y_train)
knn_model_y_pred=knn_model.predict(X_test)
print("f1 score : ",f1_score(y_test,knn_model_y_pred))
print(classification_report(y_test,knn_model_y_pred))
print(confusion_matrix(y_test,knn_model_y_pred))
```

#SVM

```python
from sklearn.svm import SVC


#linear kernel
svc_model=SVC(kernel='linear')
svc_model.fit(X_train,y_train)
svc_model_y_pred=svc_model.predict(X_test)
print("f1 score : ",f1_score(y_test,svc_model_y_pred))
print(classification_report(y_test,svc_model_y_pred))
print(confusion_matrix(y_test,svc_model_y_pred))
```

```python
#sigmoid kernel

svc_model=SVC(kernel='sigmoid')
svc_model.fit(X_train,y_train)
svc_model_y_pred=svc_model.predict(X_test)
print("f1 score : ",f1_score(y_test,svc_model_y_pred))
print(classification_report(y_test,svc_model_y_pred))
print(confusion_matrix(y_test,svc_model_y_pred))
```

```python
#rbf kernel

svc_model=SVC(kernel='rbf')
svc_model.fit(X_train,y_train)
svc_model_y_pred=svc_model.predict(X_test)
print("f1 score : ",f1_score(y_test,svc_model_y_pred))
print(classification_report(y_test,svc_model_y_pred))
print(confusion_matrix(y_test,svc_model_y_pred))
```

#DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier

dtc_model=DecisionTreeClassifier()
dtc_model.fit(X_train,y_train)
dtc_model_y_pred=dtc_model.predict(X_test)
print("f1 score : ",f1_score(y_test,dtc_model_y_pred))
print(classification_report(y_test,dtc_model_y_pred))
print(confusion_matrix(y_test,dtc_model_y_pred))
```

#RANDOM FOREST ALGORITHM

```python
from sklearn.ensemble import RandomForestClassifier

rfc_model=RandomForestClassifier()
rfc_model.fit(X_train,y_train)
rfc_model_y_pred=rfc_model.predict(X_test)
print("f1 score : ",f1_score(y_test,rfc_model_y_pred))
print(classification_report(y_test,rfc_model_y_pred))
print(confusion_matrix(y_test,rfc_model_y_pred))
```

```python
f1_score_df=pd.DataFrame({'Model':['LogisticRegression','KNeighborsClassifier','SVC','DecisionTreeClassifier','RandomForestClassifier'],
                          'f1 score':[f1_score(y_test,lr_model_y_pred),
                                      f1_score(y_test,knn_model_y_pred),
                                      f1_score(y_test,svc_model_y_pred),
                                      f1_score(y_test,dtc_model_y_pred),
                                      f1_score(y_test,rfc_model_y_pred),]})
f1_score_df
```
